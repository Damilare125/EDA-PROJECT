```python
df["Attrition"] = df["Attrition"].astype("category").cat.codes
df["BusinessTravel"] = df["BusinessTravel"].astype("category").cat.codes
df["Department"] = df["Department"].astype("category").cat.codes
df["EducationField"] = df["EducationField"].astype("category").cat.codes
df["Gender"] = df["Gender"].astype("category").cat.codes
df["JobRole"] = df["JobRole"].astype("category").cat.codes
df["MaritalStatus"] = df["MaritalStatus"].astype("category").cat.codes
df["Over18"] = df["Over18"].astype("category").cat.codes
df["OverTime"] = df["OverTime"].astype("category").cat.codes
```

```python
corr = df.corr("pearson")
corr
```

```python
# Getting the correlations of the Attrition dataset
corr = df.corr("pearson")
# Generate heatmap of correlations
plt.figure(figsize=(14,12))
sns.heatmap(corr, linewidth=1)
plt.title("CORRELATIONS HEATMAP FOR IBM ATTRITION DATASET", fontdict={"fontsize": 20})
plt.show()
```

**EXPLANATION:**
On the graph above where we can see the correlations heatmap of attrition record where JOB LEVEL, MONTHLY INCOME and TOTAL WORKING YEARS are of higher activities showing them as the 3 main factors of attrition.

**QUESTION 2**

• What is the relationship between Education and Monthly Income?

```python
# Creating a boxplot for monthly income by education.
plt.figure(figsize=(10,8))
sns.boxplot(data=df, y='MonthlyIncome', x='Education', palette="Accent") 
plt.show()
```

**EXPLANATION:**
The boxplot shows the relationship between Education and Monthly Income where we can see that the number of education certificate the employees have can determine the employees monthly income, on the graph employees with 5 certificates get more income than the employees with 1-4 education certificates and they have no outliers unlike the ones with 1-4 education certificates. 

**QUESTION 3**

• What is the effect of age on attrition?

```python
# Creating a boxplot of age by attrition.
plt.figure(figsize=(10,8))
sns.boxplot(x='Attrition', y='Age', data=df, palette='Accent')
plt.show()
```

**EXPLANATION**: The boxplot shows the effect of age on attrition whereby showing that most of the employees that left the company are 40years of age below and only few of the employees get to the retirement age between 40-60years showing them as the outliers on the graph.

**QUESTION 5**

• How does work-life balance impact the overall attrition rate?

```python
# Creating a violin plot of work life balance by attrition.
plt.figure(figsize=(10,8))
sns.violinplot(data=df, y=df['WorkLifeBalance'], x=df['Attrition'])
plt.show()
```

**EXPLANATION-** Work Life Balance impact the overall attrition because on the graph above we can see most of the employees try to balance their work life, where we can see the median to be 3 out of 4 on a scale of 1-4 which those that have 4 have a perfect work life balance and employees who are not getting the appropriate work life balance leave the work.
